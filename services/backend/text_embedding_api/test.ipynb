```python
import requests
import json

query="What were the comments by the coaches?"
```

```python
EMBEDDING_URL = 'http://localhost:8020/embed_multiple'

headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json'
}
data = json.dumps([query])

response = requests.post(EMBEDDING_URL, headers=headers, data=data)
response = response.json()

len(response["embeddings"][0])
```
